```python
%matplotlib inline

import imageio
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

imgshapes = defaultdict(int)
imgerrors = defaultdict(int)

for i in range(1617254):
    if i % 1000 == 0:
        print(i)
    try:
        img = imageio.imread("dataset/" + str(i+1) + ".png")
    except Exception as e:
        imgerrors[str(e)] += 1
    else:
        imgshapes[img.shape] += 1

print(imgshapes)
```

```python
import cv2
import h5py

BLOCKSIZE = 1000
CHUNKSIZE = 10000 # should be a multiple of the BLOCKSIZE for efficiency
IMGSHAPE = (28, 28, 4)
TOTAL = 1617254

try:
    with h5py.File('dataset.hdf5', 'x') as f:
        dset = f.create_dataset("twitch_emotes", (0,) + IMGSHAPE, dtype='uint8', chunks=(BLOCKSIZE,) + IMGSHAPE, maxshape=(None,) + IMGSHAPE, compression="gzip", shuffle=True)
except OSError:
    pass

i = 270001
while i <= TOTAL:
    image_chunk = []
    while len(image_chunk) < CHUNKSIZE and i <= TOTAL:
        try:
            img = cv2.imread("dataset/" + str(i+1) + ".png", cv2.IMREAD_UNCHANGED); # load 'as is', don't convert to bgr
            if img.shape == IMGSHAPE:
                image_chunk.append(img)
        except Exception as e:
            pass
        i += 1
    with h5py.File('dataset.hdf5', 'r+') as f:
        dset = f["twitch_emotes"]
        dset.resize(dset.shape[0] + len(image_chunk), axis=0)
        dset[-len(image_chunk):] = image_chunk
        print(dset.shape)
```
